# Pinout Microzed board
Following information guides through the connection of Pmod devices the Microzed board.

## Pmod

### Connection guide

### Pinout of Pmod

Pmod is a interface specification to connect peripheral modules to host controller boards created by Digilent. The following image shows the pinout of a typical Pmod connector, as shown on Digis Website. More information about the Digilent Pmod interface specification can be found in the [pynq documentation](http://pynq.readthedocs.io/en/latest/pynq_libraries/pmod.html?highlight=pmod) or in [this document](http://digilentinc.com/Pmods/Digilent-Pmod_%20Interface_Specification.pdf) whis if offered by Digilent.

<img width="300" src="data/pynq-z1-pmod.png">

Since, in contrast to the Pynq-Z1 board, the Microzed board does not provide dedicated Pmod connectors, it is necessary to wire them manually. The following tables show the corresponding pins of Pmod A and Pmod B on the Microzed and the Microzed Breakout Carrier Card. 

This should help to connect any Pmod devices interfaced by the Pynq software to the Microzed board.

### Wire Pmod Connectors on Microzed Board

The following images show how to connect Pmod connectors interfaced by Pynq overlays with a breadboard and Digilents [Pmod DIP: DIP to 12-pin Pmod Adapter](http://store.digilentinc.com/pmod-dip-dip-to-12-pin-pmod-adapter/) using the Microzed board. This gives a guideline on how to interface hardware connected to Pmod A or Pmod B.
